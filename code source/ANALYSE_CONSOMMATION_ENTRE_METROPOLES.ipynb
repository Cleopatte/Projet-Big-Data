```python
# créons une nouvelle colonne "communes" dans data qui contient le nombre de communes de la métropole pour chaque ligne

for city in data_commune:
    data.loc[data['Métropole'] == city, 'Communes'] = data_commune[city]

#data.head(15000)
```

```python
# créons une nouvelle colonne "Population" dans data qui contient la population de la métropole pour chaque ligne

for people in data_Population:
    data.loc[data['Métropole'] == people, 'Population(2019)'] = data_Population[people]

#data.head(15000)
```

```python
# créons une nouvelle colonne "Consommation par commune (MW)"

data['Consommation par commune (MW)'] = data['Consommation (MW)'] / data['Communes']

#data.head(15000)
```

```python
# créons une nouvelle colonne "Consommation en fonction de la population (MW)"

#data['Consommation par habitant (MW)'] = data['Consommation (MW)'] / data['Population']

#data.head(15000)
```

```python
# Nos données vont de quelle année à quelle année ?

data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month

print( data['Year'].sort_values(ascending = True).unique() )
```

```python
# Consommation globale des métropoles entre 2017 et 2023

conso_globale_par_metropole = data[['Métropole', 'Consommation (MW)']].groupby('Métropole').agg(sum).sort_values('Consommation (MW)', ascending = False).head(30)

conso_globale_par_metropole
```

```python
# histogramme de la consommation globale selon la métropole entre 2017 et 2023

fig = plt.figure(figsize=(7,5))
ax1 = fig.add_subplot(111)

ax1.bar( conso_globale_par_metropole.index , conso_globale_par_metropole['Consommation (MW)'] )
plt.xticks(rotation=80)
plt.title("Consommation globale selon la métropole")
plt.show()
```

```python
# Consommation moyenne des métropoles entre 2017 et 2023

conso_moyenne_globale_par_metropole = data[['Métropole', 'Consommation (MW)']].groupby('Métropole').mean().sort_values('Consommation (MW)', ascending = False).head(30)

conso_moyenne_globale_par_metropole
```

```python
data_2019 = data[data['Year'] == 2019]

data_2019
```

```python
# Consommation moyenne d'éléctricté de la population en fonction des métropoles en 2019

conso_moyenne_par_population_metropole_2019 = data_2019[['Population(2019)', 'Consommation (MW)']].groupby('Population(2019)').mean().sort_values('Consommation (MW)', ascending = False).head(30)

conso_moyenne_par_population_metropole_2019
```

```python
# histogramme de la consommation d'électricité moyenne de la population en fonction des métropoles

fig = plt.figure(figsize=(7,5))
ax1 = fig.add_subplot(111)

x = conso_moyenne_par_population_metropole_2019.index
y = conso_moyenne_par_population_metropole_2019['Consommation (MW)']
m, b = np.polyfit(x, y, 1) # Trouver les coefficients de la droite de régression y=ax+b

ax1.scatter(x, y)
ax1.plot(x, m*x + b, '-g') # '-r' signifie une droite rouge
plt.xticks(rotation=80)
plt.title("Consommation moyenne d'électricité de la population en 2019")
plt.show()
```
